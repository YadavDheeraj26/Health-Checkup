```python
import pandas as pd 

df = pd.read_csv("../dataset/heart.csv")
```

```python
df.head()
```

```python
target  = df.target

input = df.drop('target',axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(input,target,test_size=0.30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
x_train.head()
```

```python
col = [63,1,3,145,233,1,0,150,0,2.3,0,0,1]
pred =  model.predict(col)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

cm = confusion_matrix(pred,y_test)
```

```python
import seaborn as sns

sns.heatmap(cm,annot=True)
```

```python
import pickle

with open('./dump_model/model_heart.pkl','wb') as file:
    pickle.dump(model,file) 
```
